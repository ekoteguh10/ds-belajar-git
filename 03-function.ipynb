```python
# Import library yang dibutuhkan
import requests
```

### Task 1: Membuat function dengan parameter input `id` yang bisa mendapatkan data post berdasarkan `id` tertentu.

```python
# Fungsi untuk mengambil data post dengan id tertentu

def retrieve_data(id):
  """retrive data from specified url with certain id"""
  response = requests.get(f'https://jsonplaceholder.typicode.com/posts/{id}')
  data = response.json()
  return data
```

### Task 2: Mengambil post dengan `id` = 11 sampai dengan `id = 50` dengan menggunakan `map function`

```python
# Definisikan id yang akan diambil, misalkan selected_id dengan bantuan range(11, 51), ingat, id=51 tidak akan diambil

selected_id = range(11,51)
selected_posts = list(map(retrieve_data, selected_id))
```
